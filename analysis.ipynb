Analyze student academic performance and identify factors affecting scores, learning patterns, and improvement opportunities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data=pd.read_csv('student_data.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
data.dropna()
```

```python
data.isnull().sum()
```

```python
sns.histplot(data['G1'])
sns.histplot(data['G2'])
sns.histplot(data['G3'])
```

```python
sns.countplot(x='sex', data=data)
```

```python
sns.countplot(x='age', data=data)

```

```python
data['average_score']=(data['G1']+data['G2']+data['G3'])/3
```

```python
print(data['average_score'].head(15))
```

```python
print(data.head(15))
```

```python
data.sort_values(by='average_score', ascending=False)
```

```python
sns.boxplot(data=data[['G1','G2','G3']])

```

```python
sns.boxplot(x='sex', y='average_score', data=data)
plt.savefig('images/boxplot_of_scores.png')
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['sex'], bins=20)
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.savefig('images/distribution_of_sex.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['age'], bins=20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig('images/distribution_of_age.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['Mjob'], bins=20)
plt.title("Distribution of Mother's Job")
plt.xlabel("Mother's Job")
plt.ylabel("Frequency")
plt.savefig('images/distribution_of_mother_job.png')
plt.show()
```

```python

plt.plot(data['age'], data['absences'])

plt.title("Age vs Absences")
plt.xlabel("Age")
plt.ylabel("Absences")
plt.savefig('images/age_vs_absences.png')
plt.show()
```

GROUPING AVG SCORE BY AGE 

```python
avg_score_by_age = data.groupby('age')['average_score'].mean()
plt.plot(avg_score_by_age.index, avg_score_by_age.values)
plt.title("Average Score by Age")
```

```python
plt.bar(avg_score_by_age.index, avg_score_by_age.values)
plt.xlabel("Age")
plt.ylabel("Average Score")
plt.savefig('images/average_score_by_age.png')
```

```python
mother_job_group=data.groupby('Mjob')['average_score'].mean()
plt.bar(mother_job_group.index, mother_job_group.values)
plt.xlabel("Mother's Job")
plt.ylabel("Average Score")
plt.savefig("images/average_score_by_mother_job.png")
plt.show()
```
